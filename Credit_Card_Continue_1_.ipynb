# **Point Byserial Multicorrelation**

```python
from scipy import stats
```

```python
columns= new_df.columns
columns
```

```python
unique=new_df.nunique().reset_index()
unique["index"][0]
```

```python
#Dichotomous columns
di_col=[]
for i in range(len(columns)):
  if unique[0][i]==2:
    di_col.append(unique["index"][i])

print(di_col)
```

```python
bi_corr=np.zeros((len(di_col),len(multi_df.columns)))
p_val=np.zeros((len(di_col),len(multi_df.columns)))
```

```python
for i in range(len(di_col)):
  biserial=multi_df.corrwith(new_df[di_col[i]].astype('float'),method=stats.pointbiserialr)
  bi_corr[i]=biserial.iloc[0]
  p_val[i]=biserial.iloc[1]
```

```python
bi_corr=pd.DataFrame(bi_corr)
bi_corr.columns
```

```python
bi_corr.columns=multi_df.columns
bi_corr.index=di_col
bi_corr
```

```python
matplotlib.rcParams['figure.figsize'] = (20, 20)
sns.heatmap(bi_corr,annot=True)
```

```python
p_val=pd.DataFrame(p_val)
p_val.columns=multi_df.columns
p_val.index=di_col
p_val
```

# **Decision Tree Classifier**

Decision Tree Classifer doesn't work well if a data is unbalanced. So, I'll use SMOTE() in data having only numerical columns and RandomSample() in data containing categorical values also.

```python
#splitting dataset into X(explanatory) and Y(target variable)
X = final_df.drop(["TARGET"],axis=1)
y = final_df['TARGET']
```

```python
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from sklearn.metrics import accuracy_score, classification_report,plot_confusion_matrix
from collections import Counter
from sklearn.tree import DecisionTreeClassifier
```

```python
# split X and y into training and testing sets (SMOTE)
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0,stratify=y)
X_train.shape, X_test.shape
```

```python
print(Counter(y_train))
```

## **SMOTE**

```python
#splitting dataset into X(explanatory) and Y(target variable)
oversample = SMOTE()
X_train_smote, y_train_smote = oversample.fit_resample(X_train, y_train)
```

```python
Counter(y_train_smote)
```

```python
clf_model=DecisionTreeClassifier()
```

```python
clf_model.fit(X_train_smote,y_train_smote)
y_predict_smote=clf_model.predict(X_test)
```

```python
accuracy_score(y_test,y_predict_smote)
```

```python
matplotlib.rcParams['figure.figsize'] = (10, 10)
plot_confusion_matrix(clf_model,X_test,y_test)
```

## **Random Over Sampling**

```python
# define oversampling strategy
oversample = RandomOverSampler(sampling_strategy='minority')
# fit and apply the transform
X_train_over, y_train_over = oversample.fit_resample(X_train, y_train)
```

```python
Counter(y_train_over)
```

```python
clf_model=DecisionTreeClassifier()
```

```python
clf_model.fit(X_train_over,y_train_over)
y_predict_over=clf_model.predict(X_test)
```

```python
accuracy_score(y_test,y_predict_over)
```

```python
plot_confusion_matrix(clf_model,X_test,y_test)
```

## **Random Under Sampling**

```python
# define oversampling strategy
undersample = RandomUnderSampler(sampling_strategy='majority')
# fit and apply the transform
X_train_under, y_train_under = undersample.fit_resample(X_train, y_train)
```

```python
Counter(y_train_under)
```

```python
clf_model=DecisionTreeClassifier()
```

```python
clf_model.fit(X_train_under,y_train_under)
y_predict_under=clf_model.predict(X_test)
```

```python
accuracy_score(y_test,y_predict_under)
```

```python
plot_confusion_matrix(clf_model,X_test,y_test)
```
